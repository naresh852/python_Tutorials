######   Import required libraries

```python
import os #To read files
import pickle #To store image features
import numpy as np
from tqdm.notebook import tqdm  #to create progress bar for a loop

from tensorflow.keras.applications.vgg16 import VGG16, preprocess_input # extracting features and preprocessing image data for vgg
from tensorflow.keras.preprocessing.image import load_img, img_to_array #preprocessing the image
from tensorflow.keras.preprocessing.text import Tokenizer #preprocessing the text
from tensorflow.keras.preprocessing.sequence import pad_sequences #even out the words suppose 10 words in 1 row, 5 words in 2 row fill 0's for remaining nulls
from tensorflow.keras.models import Model #models 
from tensorflow.keras.utils import to_categorical, plot_model #convert to matrix  binary class and more utilities to see Architecture of model
from tensorflow.keras.layers import Input, Dense, LSTM, Embedding, Dropout, add#Layers basic building blocks of neural networks 
```

###### Extract Image features

```python
#load vgg16 model
model = VGG16()
#restructure the model
model = Model(inputs = model.inputs, outputs = model.layers[-2].output) #leaving the last layers
#summarise
print(model.summary())
```

```python
#Extract features from Image
features = {}
directory = os.path.join(BASE_DIR, 'Images')

for img_name in tqdm(os.listdir(directory)):
    #load the image from the file
    img_path = directory +'/'+ img_name
    image = load_img(img_path, target_size=(224,224))
    
    #convert image pixels to numpy array
    image = img_to_array(image)
 
    #reshape data for model in order to extract features
    image = image.reshape((1, image.shape[0], image.shape[1], image.shape[2]))
     
    #prepocess image for vgg
    image = preprocess_input(image)
    #extract features 
    feature = model.predict(image, verbose=0) #verbose=0 no to display unwanted lines of data
    
    #get image ID
    image_id = img_name.split('.')[0] #to split .jpg files
    print(image_id)
    #store features
    features[image_id] = feature
    
```

```python
#store features in a pickle
pickle.dump(features, open(os.path.join(WORKING_DIR, 'features.pkl'),'wb'))
```

###### Load Captions Data

```python
with open(os.path.join(BASE_DIR, 'captions\\captions.txt'), 'r') as f:
    next(f) #to skip the heading line
    captions_doc = f.read() #reads the whole text data
    
```

```python
#create mapping of image to captions
mapping = {}
#process lines
for line in tqdm(captions_doc.split('\n')):
    #split the line by comma(,)
    tokens = line.split(',')
    if len(line) < 2:
        continue
    image_id, caption = tokens[0], tokens[1:]
    #remove extension b.jpg from image id
    image_id = image_id.split('.')[0]
    #convert captions list to string
    caption = " ".join(caption)
    #create list if needed
    if image_id not in mapping:
        mapping[image_id]=[]
    #store the caption
    mapping[image_id].append(caption)
```

```python
len(mapping)
```

###### Preprocess caption data

```python
def clean(mapping):
    for key, captions in mapping.items():
        for i in range(len(captions)):
            #take one caption at a time
            caption = captions[i]
            #preprocessing steps
            #convert to lowercase
            caption = caption.lower()
            #delete digits, special chars, etc.,
            caption = caption.replace('[^A-Za-z]','')
            #delete additional spaces
            caption = caption.replace('\s+', ' ')
            #add start and end tags to the caption
            caption = '<start> '+" ".join([word for word in caption.split() if len(word)>1])+' <end>'
            captions[i] = caption
```

###### Train Test Split

```python
image_ids = list(mapping.keys())
split = int(len(image_ids) * 0.90)
train = image_ids[:split]
test = image_ids[split:]
print(len(train))
```

```python
#create data generator to get data in batch (avoid session crash)
def data_generator(data_keys, mapping, features, tokenizer, max_length, vocab_size, batch_size):
    #loop over images
    X1, X2, y = list(), list(), list()
    n=0
    while 1:
        for key in data_keys:
            print("printing keys"+str(key))
            n + = 1
            print("printing captions")
            captions = mapping[key]
            print(captions)
            #process each caption
            for caption in captions:
                print("single caption")
                print(caption)
                #encode the sequence
                seq = tokenizer.texts_to_sequences([caption])[0] #texts_to_sequences assign integers to words. For example, ‘machine’ is assigned value 2.
                print("sequence and its length")
                print(seq)
                print(len(seq))
                #split the sequence into X, y pairs
                for i in range(1, len(seq)):
                    #split into input and output pairs
                    in_seq, out_seq = seq[:i], seq[i]
                    print("printing input seq")
                    print(in_seq)
                    print("printing output seq")
                    print(out_seq)
                    #pad input sequence
                    in_seq = pad_sequence([in_seq], maxlen = max_length)[0]
                    print("print padding input seq")
                    print(in_seq)
                    #encode output sequence
                    out_seq = to_categorical([out_seq]), num_classes = vocab_size)[0] #converting to one hot encoding
                    print("print encode categorical out_seq")
                    print(out_seq)
                    
                    #store the sequences
                    X1.append(features[key][0])
                    print("features key")
                    print(features[key])
                    print(features[key][0])
                    X2.append(in_seq)
                    print("final in_seq")
                    print(in_seq)
                    y.append(out_seq)
                    print("out_seq")
                    print(out_seq)
                if n == batch_size:
                    X1, X2, y = np.array(X1), np.array(X2), np.array(y)
                    yield[X1, X2], y
                    X1, X2, y = list(), list(), list()
                    n=0
                    
```

```python
#encoder model
#Image feature layers
inputs1 =  Input(shape=(4096,))
fe1 = Dropout(0.4)(inputs1) #Droput regularisation approach to prevent over fitting
fe2 = Dense(256,activation='relu')(fe1) #Dense current neuron layer gets its input from previous neuron layer
#relu rectified linear unit is a activation function which is non - linear with values 0.0 to positive value

#Sequence feature layers
inputs2 = Input(shape=(max_length,))
se1 = Embedding(vocab_size, 256,mask_zero=True)(inputs2) #dimesions of the input mask_zero = true to indicate padding is done please ignore th zero's
#Embedding is grouping of similar semantic  items together go,let'sgo, going as "GO"
se2 = Dropout(0.4)(se1)
se3 = LSTM(256)(se2) #LSTM is a RNN

#decode model #we are going to concatnate image and text features
decoder1 = add([fe2,se3])
decoder2 = Dense(256,activation='relu')(decoder1)
outputs = Dense(vocab_size, activation='softmax')(decoder2) #softmax is used as a last activation function for multi class neural networks
#to normalize the output of a network to a probability distribution over predicted output classes
#Sigmoid is used for binary classification methods where we only have 2 classes, while SoftMax applies to multiclass problems. 
#In fact, the SoftMax function is an extension of the Sigmoid function.

model = Model(inputs=[inputs1, inputs2],outputs=outputs)
model.compile(loss = 'categorical_crossentropy', optimizer='adam')
#binary crossentropy ( binary_crossentropy ) is used when you have a two-class, or binary, classification problem. categorical crossentropy ( categorical_crossentropy ) is used for a multi-class classification problem. mean squared error ( mean_squared_error ) is used for a regression problem.
#A loss function is a function that compares the target and predicted output values lesser the loss higher the model accuracy
#sparse entrophy is used when labels are integers (0,1,2,.....)
#ADAM This optimization algorithm is a further extension of stochastic gradient descent to update network weights during training. Unlike maintaining a single learning rate through training in SGD, Adam optimizer updates the learning rate for each network weight individually.

#plot the model
plot_model(model, show_shapes=True)
```

###### Train the model


```python
#Stochastic gradient descent is an iterative learning algorithm that uses a training dataset to update a model.
#The batch size is a hyperparameter of gradient descent that controls the number of training samples to work through before the model’s internal parameters are updated.
#The number of epochs is a hyperparameter of gradient descent that controls the number of complete passes through the training dataset.
```

```python
epochs = 15 #An epoch in a neural network is the training of the neural network with all the training data for one cycle.
batch_size = 64
steps = len(train)//batch_size

for i in range(epochs):
#create data generator
	generator = data_generator(train, mapping, features, tokenizer, max_length, vocab_size, batch_size)
	#fit for one epoch
    model.fit(generator, epochs=1, steps_per_epoch= steps, verbose=1 )
    #When verbose is turned on, the algorithm will provide more detailed information about its progress as your model iterates through the training process
```

```python
#save the model
model.save(WORKING_DIR+'/best_model.h5')
```

###### Generate captions for Image

```python
def idx_to_word(integer, tokenizer):
    for word, index in tokenizer.word_index.items():
        if index == integer:
            return word
    return None
```

```python
#generate caption for an image
def predict_caption(model, image, tokenizer, max_length):
    in_text = '<start>'
    #iterate over the max length of sequence
    for i in range(max_length):
        #encode input sequence
        sequence = tokenizer.texts_to_sequence([in_text])[0]
        #pad the sequence
        sequence = pad_sequences([sequence], max_length)
        #predict next word
        yhat = model.predict([image, sequence], verbose=0)
        #get index with high probability
        yhat = np.argmax(yhat) #It's somewhat similar to the Numpy maximum function, but instead of returning the maximum value, it returns the index of the maximum value
        #convert index to word
        word = idx_to_word(yhat, tokenizer)
        #stop if word not found
        if word is None:
            break
        #append word as input for generating next word
        in_text +=" "+word
        #stop if we reach end tag
        if word == '<end>':
            break
    return in_text
```

```python
from nltk.transalate.bleu_score import corpus_bleu
#validate with text data
actual, predicted = list(), list()

for key in tqdm(test):
    #get actual caption
    captions = mapping[key]
    #predict the caption for image
    y_pred = predict_caption(model, features[key], tokenizer, max_length)
    #split into words
    actual_captions = [caption.split() for caption in captions]
    #append to the list
    actual.append(actual_captions)
    
    predicted.append(y_pred.split())
    
#calculate BLEU score
print("BLEU-1: %f"%corpus_bleu(actual, predicted, weights = (1.0,0,0,0)))
print("BLEU-2: %f"%corpus_bleu(actual, predicted, weights = (0.5,0.5,0,0)))
```
